```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load CSV file
df = pd.read_csv('data.csv')  # Replace 'data.csv' with your filename

#Extract data from experiments
d0_data = df.iloc[0:29, 0:]
d5_data = df.iloc[29:44, 0:]
d10_data = df.iloc[44:59, 0:]
s0_data = df.iloc[59:73, 0:]
s5_data = df.iloc[73:87, 0:]
s10_data = df.iloc[87:, 0:]

plt.rcParams['legend.title_fontsize'] = 8
```

```python
def plot_sigbkgtot(data):

    plt.figure(dpi=(250))

    xd = list(data.iloc[0:,0])
    xd = [round(x,2) + 3.66 for x in xd]

    xs = list(data.iloc[0:,1])
    xs = [round(x,2) + 1.10 + 0.6 for x in xs]

    x = xd
    x_err = list(data.iloc[0:,9])
    x_label = r'$x_{d}$ (cm)'
    title = r'Count Rate vs Detector-to-Foil Distance ($x_{s}$ = ' + str(round(xs[0],2)) + 'cm)'

    if xd[0] == xd[1]:
        x = xs
        x_err = list(data.iloc[0:,10])
        x_label = r'$x_{s}$ (cm)'
        title = r'Count Rate vs Source-to-Foil Distance ($x_{d}$ = ' + str(xd[0]) + 'cm)'
    
    sig = list(data.iloc[0:,8])
    bkg = list(data.iloc[0:,7])
    tot = list(data.iloc[0:,6])

    sig_err = list(data.iloc[0:,13])
    bkg_err = list(data.iloc[0:,12])
    tot_err = list(data.iloc[0:,11])

    # Create scatter plot with error bars
    plt.errorbar(x, bkg, xerr=x_err, yerr=bkg_err, fmt='none',  lw = 1, color='red', alpha=0.5, label='Background')
    plt.errorbar(x, tot, xerr=x_err, yerr=tot_err, fmt='none', lw = 1, color='blue', alpha=0.5, label='Total')
    plt.errorbar(x, sig, xerr=x_err, yerr=sig_err, fmt='none', lw = 1, color='green', alpha=0.5, label='Signal')

    # plt.plot(x,bkg, color = 'red', linestyle='dotted')
    # plt.plot(x,tot, color='blue', linestyle='dotted')
    #plt.plot(x,sig, color ='green', linestyle='dotted')

    # Labels and legend
    plt.xlabel(x_label, fontsize=10)
    plt.ylabel('Count Rate (Hz)', fontsize=10)
    plt.title(title, fontsize=10)
    plt.legend(fontsize=8)

    plt.show()
```
